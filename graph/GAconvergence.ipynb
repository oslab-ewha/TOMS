```python
# GA 수렴 그래프

import os
import pandas as pd
import matplotlib.pyplot as plt
import shutil

file_path = "report.txt"
data = pd.read_csv(file_path, sep=" ", comment="#", header=None, names=[
    "generation", "power_min", "power_avg", "power_max", "util_min", "util_avg", "util_max"
])

# x축 1500까지
limited_data_one = data[data["generation"] <= 1500]

# Cost function
plt.figure(figsize=(10, 6))
plt.plot(limited_data_one["generation"], limited_data_one["power_min"], label="Best", linewidth=2.5)
plt.plot(limited_data_one["generation"], limited_data_one["power_avg"], label="Average", linewidth=2.5)
plt.plot(limited_data_one["generation"], limited_data_one["power_max"], label="Worst", linewidth=2.5)
plt.xlabel("Generation", fontsize=25)
plt.ylabel("Cost function", fontsize=25)
plt.legend(fontsize=20, frameon=True, edgecolor="black", loc="upper right", bbox_to_anchor=(1, 1), borderaxespad=0.3)

plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.grid()
plt.savefig("GA_convergence_CostFunction.png", bbox_inches="tight")
plt.show()
plt.close()

# Utilization
plt.figure(figsize=(10, 6))
plt.plot(limited_data_one["generation"], limited_data_one["util_min"], label="Worst", linewidth=2.5)
plt.plot(limited_data_one["generation"], limited_data_one["util_avg"], label="Average", linewidth=2.5)
plt.plot(limited_data_one["generation"], limited_data_one["util_max"], label="Best", linewidth=2.5)
plt.xlabel("Generation", fontsize=25)
plt.ylabel("Utilization", fontsize=25)
plt.legend(fontsize=20, frameon=True, edgecolor="black", loc="upper right", bbox_to_anchor=(1, 1), borderaxespad=0.3)

plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.grid()
plt.savefig("GA_convergence_Utilization.png", bbox_inches="tight")

plt.show()
plt.close()
```
